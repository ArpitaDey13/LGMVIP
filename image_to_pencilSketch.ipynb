# Import the necessary libraries

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

# Import the image and displaying the BGR image

```python
img = cv2.imread("cat_image.jpg")
plt.imshow(img)
plt.axis('off')
# the photo is inconsistent because matplotlib uses RGB color scheme and openCV uses BGR color scheme
```

#  Displaying the RGB image

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
```

# Convert Image to pencil sketch

## 1. Convert to gray image

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
plt.axis(False)
```

## 2. Invert the gray image

```python
invert_img=cv2.bitwise_not(grey_img)
plt.imshow(invert_img)
plt.axis(False)
```

## 3.Blur the inverted image

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
plt.imshow(blur_img)
plt.axis(False)
```

## 4. Invert the blurred image

```python
invblur_img=cv2.bitwise_not(blur_img)
plt.imshow(invblur_img)
plt.axis(False)
```

## 5. Constructing the sketch image

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
plt.imshow(sketch_img)
plt.axis(False)
```

## Saving the sketch

```python
cv2.imwrite("sketch.png", sketch_img)
```

## Comparing both images

```python
plt.subplot(1,2,1)
plt.title('Original', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
